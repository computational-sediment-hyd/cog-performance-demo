## make VRT file from all the .tif files


```sh
gdalbuildvrt -srcnodata -9999 -vrtnodata -9999 -a_srs EPSG:6676 photo.vrt original/*.tif
```

````python
from osgeo import gdal 
import glob
import os

fpath = r"original"
g = glob.glob( os.path.join(fpath, '*.tif') )

gdal.UseExceptions()

d = gdal.BuildVRT(
    'photo.vrt',
    g,
    srcNodata=-9999,
    VRTNodata=-9999,
    outputSRS='EPSG:6676'  # Use outputSRS instead of CRS
)
d = None
````

## VRT file to geotiff : no compression

```sh
gdal_translate -of GTiff photo.vrt photo.geotiff
```

````python
from osgeo import gdal 

gdal.UseExceptions()
# Set translation options
translate_options = gdal.TranslateOptions(
    format="GTiff")

# Run gdal_translate
d = gdal.Translate(
    destName='photo.geotiff',
    srcDS='photo.vrt',
    options=translate_options
)
d = None
````

## VRT file to geotiff : compression

```sh
gdal_translate -of GTiff -co COMPRESS=JPEG -co QUALITY=85 -co PHOTOMETRIC=YCBCR photo.vrt photo_compressed.geotiff
```


````python
from osgeo import gdal 

gdal.UseExceptions()
# Set translation options
translate_options = gdal.TranslateOptions(
    format="GTiff",
    creationOptions=[
        "COMPRESS=JPEG",
        "QUALITY=85",
        "PHOTOMETRIC=YCBCR"
    ]
)

# Run gdal_translate
d = gdal.Translate(
    destName='photo_compressed.geotiff',
    srcDS='photo.vrt',
    options=translate_options
)
d = None
````
